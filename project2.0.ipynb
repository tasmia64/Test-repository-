```python
import pandas as pd

# Load the CSV file
file_path = "D:/DSP/PAH_Samples.csv"
data = pd.read_csv(file_path)
```

```python
# 1. Handling Missing Values
# Separate numeric and categorical columns
numeric_cols = data.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = data.select_dtypes(include=['object']).columns

# Fill missing values
data_cleaned = data.copy()
data_cleaned[numeric_cols] = data_cleaned[numeric_cols].fillna(data_cleaned[numeric_cols].mean())
data_cleaned[categorical_cols] = data_cleaned[categorical_cols].fillna("Unknown")

# 2. Convert Date Columns to Datetime
date_cols = ['Date_Col', 'Date_Rec', 'Date_Extrc', 'Date_Analy'] 
for col in date_cols:
    data_cleaned[col] = pd.to_datetime(data_cleaned[col], errors='coerce')

# 3. Save the cleaned data 
cleaned_file_path = 'cleaned_dataset.csv'
data_cleaned.to_csv(cleaned_file_path, index=False)

print("Data cleaning complete. Cleaned dataset saved to:", cleaned_file_path)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the cleaned dataset
cleaned_file_path = "C:/Users/Pacific BD/Downloads/Project/cleaned_dataset.csv"
data_cleaned = pd.read_csv(cleaned_file_path)

# Convert date columns (if not already done)
date_cols = ['Date_Col', 'Date_Rec', 'Date_Extrc', 'Date_Analy']
for col in date_cols:
    if col in data_cleaned.columns:
        data_cleaned[col] = pd.to_datetime(data_cleaned[col], errors='coerce')

# 1. Geographic Distribution: Latitude vs. Longitude
plt.figure(figsize=(8, 4))
sns.scatterplot(data=data_cleaned, x='Longitude', y='Latitude', hue='Region', palette='viridis', alpha=0.7)
plt.title('Geographic Distribution of Samples')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# 2. Sampling Frequency Over Time
if 'Date_Col' in data_cleaned.columns:
    plt.figure(figsize=(8, 4))
    data_cleaned['Year_Col'] = data_cleaned['Date_Col'].dt.year
    sns.countplot(data=data_cleaned, x='Year_Col', color='teal')
    plt.title('Number of Samples Collected Per Year')
    plt.xlabel('Year')
    plt.ylabel('Sample Count')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# 3. Distribution of a Chemical Concentration (e.g., 'Naph')
if 'Naph' in data_cleaned.columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(data_cleaned['Naph'], kde=True, color='blue')
    plt.title('Distribution of Naph Concentrations')
    plt.xlabel('Naph Concentration')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# 4. Correlation Heatmap for Chemical Measurements
# Select only numeric columns for correlation
chem_cols = ['Naph', 'Phen', 'DBT', 'Fluor', 'Pyr', 'Sum_PAH']  # Example chemical columns
chem_data = data_cleaned[chem_cols].dropna()

plt.figure(figsize=(10, 6))
correlation_matrix = chem_data.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap of Chemical Measurements')
plt.show()
```
